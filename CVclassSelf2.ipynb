```python
import numpy as np
import matplotlib.pyplot as plt

# Define camera intrinsic parameters
f = 800                   # Focal length in pixels
cx, cy = 320, 240         # Principal point (image center) in pixels

# Create a set of 3D points (for example, vertices of a cube)
# The cube is centered at the origin with varying depths
points_3D = np.array([[x, y, z] for x in [-1, 1]
                                 for y in [-1, 1]
                                 for z in [5, 7]])

# Project 3D points to 2D image plane using the pinhole camera model
points_2D = []
for point in points_3D:
    X, Y, Z = point
    u = f * (X / Z) + cx
    v = f * (Y / Z) + cy
    points_2D.append([u, v])
points_2D = np.array(points_2D)

# Plot the projected points on the image plane
plt.figure(figsize=(6, 4))
plt.scatter(points_2D[:, 0], points_2D[:, 1], color='red', s=50)
plt.title("Pinhole Camera Projection")
plt.xlabel("u (pixels)")
plt.ylabel("v (pixels)")
plt.xlim(0, 640)
plt.ylim(480, 0)  # Invert y-axis to match image coordinates (origin at top-left)
plt.grid(True)
plt.show()
```
